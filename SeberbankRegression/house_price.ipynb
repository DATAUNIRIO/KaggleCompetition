```python
import shutil
import os
import time
import pickle
import pandas as pd
import requests
from bs4 import BeautifulSoup
from __future__ import unicode_literals
```

```python
list_3 = pickle.load(('area_list.pkl', 'r'), f)
```

```python
for i, (rus_name, area_name, url) in enumerate(list_3):
    print i, rus_name, area_name, url
    if i <= 29:
        continue
    
    page = requests.get('http://www.irn.ru{}'.format(url))
    page_soup = BeautifulSoup(page.text, "lxml")
    link = [img.get('src') for img in page_soup.find_all('img') if img.get('src').startswith("/graph/services/index_history.php?what=IS")][0]
    geo_id = link.split('&')[-2][7:]

    os.mkdir(area_name)

    currency_list = [(0, 'rub'), (1, 'usd'), (2, 'eur')] 
    for currency_id, currecy_str in currency_list:
        url = 'http://www.irn.ru/graph/services/index_history.php?what=IS&type=geo&period=5&geo_id={}&currency={}'.format(geo_id, currency_id)
        response = requests.get(url, stream=True)
        with open('{}/{}_{}.png'.format(area_name, area_name, currecy_str), 'wb') as out_file:
            shutil.copyfileobj(response.raw, out_file)
        del response
        time.sleep(1) 

    time.sleep(3)    
```
